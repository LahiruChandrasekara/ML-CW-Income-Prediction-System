 Unique Values

```python
print(data.nunique()) # unique value of the each column
```

```python
# convert int64 to object

column_to_convert = ['age', 'fnlwgt','education-num','capital-gain','captial-loss','hours-per-week']
data[column_to_convert] = data[column_to_convert].astype(str)
```

For get a better idea about the dataset lets check the unique values for each column

```python
# Find out unique values for each column
unique_values = {}
for column in data:
    unique_values[column] = data[column].unique()

for col, val in unique_values.items():
    print(f"Unique Values for each {col} : {val}")
    print()
```

There are " ?" values in columns (workclass, occupation, native-country) so need to remove them 

```python
# get rid " ?" marks
data.replace(' ?',pd.NA, inplace= True)
```

```python
# lets check if the process is successfull
unique_values_new = {}
for column in data:
    unique_values_new[column] = data[column].unique()

for col, val in unique_values_new.items():
    print(f"Unique Values for each {col} : {val}")
    print()
```

After removing ' ?' values lets check the Null values in the dataset

```python
# Checking the Null values
print(data.isnull().sum())
```
